# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go  # pip install plotly
```

# Beale Test Function

http://www.sfu.ca/~ssurjano/beale.html

$ f(\mathbf{x}) = (1.5 - x_1 + x_1 x_2)^2 + (2.25 - x_1 + x_1 x_2^2)^2 + (2.625 - x_1 + x_1 x_2^3)^2$

```python
# ---- tweakable parameters ----
x1_min, x1_max = -4.5, 4.5      # domain limits for x1
x2_min, x2_max = -4.5, 4.5      # domain limits for x2
n_x1 = 100                      # number of points in x1 for visual
n_x2 = 100                      # number of points in x2 for visual
# --------------------------------

def test_function(x1, x2):      # Beale Test Function: http://www.sfu.ca/~ssurjano/beale.html
    return ((1.5   - x1 + x1*x2)      ** 2 +
            (2.25  - x1 + x1*x2**2)   ** 2 +
            (2.625 - x1 + x1*x2**3)   ** 2)

# grid
x1 = np.linspace(x1_min, x1_max, n_x1)
x2 = np.linspace(x2_min, x2_max, n_x2)
X1, X2 = np.meshgrid(x1, x2)
y = test_function(X1, X2)               # output surface
```

## Contour Heatmap

```python
fig, ax = plt.subplots(figsize=(6, 5))
cntr = ax.contourf(X1, X2, y, levels=50, cmap='viridis')
fig.colorbar(cntr, ax=ax)
ax.set_xlabel(r'$x_1$'); ax.set_ylabel(r'$x_2$')
ax.set_title('Beale function (Contour Heatmap)')
plt.tight_layout(); plt.show()
```

## Static 3-D Surface

```python
# TODO: adjust vertical axis to be 25, 50, 75, etc., with a *10^2 marker

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X1, X2, y, rstride=3, cstride=3, cmap='viridis',
                       linewidth=0, antialiased=True)
ax.set_xlabel(r'$x_1$'); ax.set_ylabel(r'$x_2$'); ax.set_zlabel(r'$y$')
ax.set_title('Beale function surface')
#fig.colorbar(surf, shrink=0.6, aspect=10)  # optional color bar
plt.tight_layout(); plt.show()
```

# Interactable 3-D Surface

```python
fig = go.Figure(
    data=[go.Surface(x=X1, y=X2, z=y, colorscale='Viridis', showscale=True)]
)
fig.update_layout(title='Beale function – interactive',
                  scene=dict(xaxis_title='x1',
                             yaxis_title='x2',
                             zaxis_title='y'),
                  width=750, height=600,
                  margin=dict(l=50, r=50, b=50, t=50))
fig.show()
```
